```python
import pandas as pd
import numpy as np
import datetime
import csv
import time
```

```python
from bs4 import BeautifulSoup
import urllib2
```

```python
yearmin = 2011
yearmax = 2018
```

## Scrape QB specific stats from Pro-Football-Reference

```python
yearlist = range(yearmin,yearmax)
for ii in yearlist:
    url = "https://www.pro-football-reference.com/years/%s/passing.htm" %ii
    html = urllib2.urlopen( url ).read()
    soup = BeautifulSoup(html)
    cnt = 0
    data = []
    rows = soup.find_all('th')
    for row in rows:
        data = np.append(data, row.find(text=True))
        cnt+=1
    qbnames = list(data[1:30].astype(str))
    qbnames[0] = 'QB'
    qbnames[23] = 'Sk_Yds'
    rows = soup.find_all('td')
# for jj in rows:
    cnt = 0
    data = []
    rows = soup.find_all('td')
    for row in rows:
        data = np.append(data, row.find(text=True))
        cnt+=1
    data = data.astype(str)
    col = 29
    row = len(data)/29.
    data = data.reshape((int(row),int(col)))
    df = pd.DataFrame(data, columns = qbnames)
    df['year'] = ii
    if ii == yearlist[0]:
        df_qbstat = df.copy()
    else:
        df_qbstat = pd.concat([df_qbstat,df])
    print 'the year is ' + str(ii)
    time.sleep(1)
df_qbstat['4QC'].loc[df_qbstat['4QC']=='None'] = 0
df_qbstat['GWD'].loc[df_qbstat['GWD']=='None'] = 0
df_qbstat['Y/C'].loc[df_qbstat['Y/C']=='None'] = 0
df_qbstat['QBR'].loc[df_qbstat['QBR']=='None'] = 0
df_qbstat['QBrec'].loc[df_qbstat['QBrec']=='None'] = '0-16-0'
# df_qbstat['QBrec'].loc[df_qbstat['QBrec']=='0-0-0'] = '0-16-0'
df_qbstat['Age'] = df_qbstat['Age'].astype(int)
df_qbstat['G'] = df_qbstat['G'].astype(int)
df_qbstat['GS'] = df_qbstat['GS'].astype(int)
df_qbstat['Cmp'] = df_qbstat['Cmp'].astype(int)
df_qbstat['Att'] = df_qbstat['Att'].astype(int)
df_qbstat['Cmp%'] = df_qbstat['Cmp%'].astype(float)
df_qbstat['Yds'] = df_qbstat['Yds'].astype(int)
df_qbstat['TD'] = df_qbstat['TD'].astype(int)
df_qbstat['TD%'] = df_qbstat['TD%'].astype(float)
df_qbstat['Int'] = df_qbstat['Int'].astype(int)
df_qbstat['Int%'] = df_qbstat['Int%'].astype(float)
df_qbstat['Lng'] = df_qbstat['Lng'].astype(int)
df_qbstat['Y/A'] = df_qbstat['Y/A'].astype(float)
df_qbstat['AY/A'] = df_qbstat['AY/A'].astype(float)
df_qbstat['Y/C'] = df_qbstat['Y/C'].astype(float)
df_qbstat['Rate'] = df_qbstat['Rate'].astype(float)

df_qbstat['QBR'] = df_qbstat['QBR'].astype(float)
df_qbstat['QBwin'] = df_qbstat['QBrec'].apply(lambda x: float(str(x).split('-')[0]))
df_qbstat['QBlose'] = df_qbstat['QBrec'].apply(lambda x: float(str(x).split('-')[1]))
df_qbstat['QBtie'] = df_qbstat['QBrec'].apply(lambda x: float(str(x).split('-')[2]))
df_qbstat['Sk'] = df_qbstat['Sk'].astype(int)
df_qbstat['Sk_Yds'] = df_qbstat['Sk_Yds'].astype(int)
df_qbstat['NY/A'] = df_qbstat['NY/A'].astype(float)
df_qbstat['ANY/A'] = df_qbstat['ANY/A'].astype(float)
df_qbstat['Sk%'] = df_qbstat['Sk%'].astype(float)
df_qbstat['4QC'] = df_qbstat['4QC'].astype(int)
df_qbstat['GWD'] = df_qbstat['GWD'].astype(int)
df_qbstat['QBwinP'] = df_qbstat['QBwin']/(df_qbstat['QBwin']+df_qbstat['QBlose']+df_qbstat['QBtie'])
```

## Scrape WR specific stats from Pro-Football-Reference

```python
yearlist = range(yearmin,yearmax)
wrs=[]
for ii in yearlist:
    url = "https://www.pro-football-reference.com/years/%s/receiving.htm" %ii
    html = urllib2.urlopen( url ).read()
    soup = BeautifulSoup(html)
    cnt = 0
    rows = soup.find_all('th')
    wrdat = []
    for row in rows:
        wrdat = np.append(wrdat, row.find(text=True))
    wrnames = np.append('WR',wrdat[2:17])
    
    
    data = []
    rows = soup.find_all('td')
    cnt = 0
    
    for row in rows:
        data = np.append(data, row.find(text=True))
        cnt+=1
    data = data.astype(str)
    ncol =16
    data = data.reshape(len(data)/ncol,ncol)
    df = pd.DataFrame(data, columns = wrnames)
    df['year'] = ii
    wrs = np.append(wrs,df['WR'].values)
    if ii == yearlist[0]:
        df_wrstat = df.copy()
    else:
        df_wrstat = pd.concat([df_wrstat,df])
    print 'the year is ' + str(ii)
    time.sleep(1)
df_wrstat = df_wrstat.rename(index = str, columns = {'TD':'Rec_TD','Lng':'Rec_Lng','Y/G':'Rec_Y/G','Yds':'Rec_Yds'})
```
